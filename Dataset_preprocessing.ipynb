### 1. Kickstarter Dataset Preprocessing

```python
import warnings
warnings.simplefilter("ignore")  #Don't show warning messages

import pandas as pd 
import numpy as np
pd.set_option('display.max_columns', None) #Done to displays all columns when called

KSdata = pd.read_csv("ks-projects-201801.csv")
print(KSdata.shape)
#KSdata.describe() #data statistics
```

```python
KSdata.isnull().sum() # Done to check for missing values.
```

```python
KSdata.head().transpose() # Transpose is done to show all columns
```

```python
KSdata.drop(KSdata.columns[[0, 1]], axis=1, inplace=True) #dropping ID and name column
KSdata.head().transpose()
```

```python
KSdata['usd pledged'] = KSdata['usd pledged'].fillna(KSdata['usd pledged'].mean()) #since 'usd pledge'is a numerical value, the mean is used to fill all missen values in the column
```

```python
KSdata.isnull().sum() # Done to check for missing values
```

```python
KSdata['state'].unique() #checking the values of state
```

```python
KSdata = KSdata.drop(KSdata.query("state !='failed' and state !='successful'").index, axis=0).reset_index(drop=True) #paying attention to data with either failed or successful outcome and dropping the rest
```

```python
KSdata['state'].unique()
```

```python
#extracting the year and month from deadline and launched column
KSdata['deadline_year'] = KSdata['deadline'].apply(lambda x: float(x[0:4]))
KSdata['deadline_month'] = KSdata['deadline'].apply(lambda x: float(x[5:7]))

KSdata['launched_year'] = KSdata['launched'].apply(lambda x: float(x[0:4]))
KSdata['launched_month'] = KSdata['launched'].apply(lambda x: float(x[5:7]))

KSdata = KSdata.drop(['deadline', 'launched'], axis=1) #dropping the old columns
```

```python
KSdata['state'] = KSdata['state'].apply(lambda x: 1 if x == 'successful' else 0) #changing the state column to numerical values
```

```python
KSdata.head().transpose()
```

```python
print(KSdata.apply(lambda col: col.unique())) #checking to know the unique values in each column
```

```python
KSdata = pd.concat([KSdata, pd.get_dummies(KSdata["main_category"], prefix = "main_category"), 
                      pd.get_dummies(KSdata["country"], prefix = "country")], axis = 1)

KSdata=KSdata.drop(["main_category", "country", "currency", "category"], axis = 1) #dropping columns
```

```python
KSdata.head().transpose()
```

##### Kickstarter Dataset is ready for classification

### 2. Car Insurance Dataset Preprocessing

```python
cardatatrain=pd.read_csv('carInsurance_Train.csv')
cardatatest=pd.read_csv('carInsurance_Test.csv')

print(cardatatrain.shape)
#cardatatrain.describe() #data statistics
```

```python
print(cardatatest.shape)
#cardatatest.describe() #data statistics
```

```python
cardatatrain.head().transpose() # Transpose is done to show all columns
```

```python
cardatatrain.isnull().sum() # Done to check for missing values.
```

```python
cardatatest.isnull().sum() # Done to check for missing values.
```

```python
cardatatest=cardatatest.drop(["Id", "Outcome"], axis = 1)
```

```python
cardatatrain=cardatatrain.drop(["Id", "Outcome"], axis = 1)
```

```python
# filling categorical missing values with column modes
for column in ['Job', 'Education', 'Communication']:
        cardatatrain[column] = cardatatrain[column].fillna(cardatatrain[column].mode()[0])
```

```python
# filling categorical missing values with column modes
for column in ['Job', 'Education', 'Communication']:
        cardatatest[column] = cardatatest[column].fillna(cardatatest[column].mode()[0])
```

```python
cardatatest.isnull().sum() # Done to check for missing values.
```

```python
cardatatest =cardatatest.drop(["CarInsurance"], axis = 1)
```

```python
cardatatest.isnull().sum() # Done to check for missing values.
```

```python
cardatatrain.isnull().sum() # Done to check for missing values.
```

```python
# Done to know the amount of unique values in all categorical variables
{column: len(cardatatrain[column].unique()) for column in cardatatrain.select_dtypes('object').columns}
```

```python
# Done to know the amount of unique values in all categorical variables
{column: len(cardatatest[column].unique()) for column in cardatatest.select_dtypes('object').columns}
```

```python
#extracting duration from CallEnd and CallStart times
cardatatrain['CallDuration'] = (pd.to_datetime(cardatatrain['CallEnd']) - pd.to_datetime(cardatatrain['CallStart'])).apply(lambda x: x.seconds)
```

```python
cardatatest['CallDuration'] = (pd.to_datetime(cardatatest['CallEnd']) - pd.to_datetime(cardatatest['CallStart'])).apply(lambda x: x.seconds)
```

```python
cardatatrain = cardatatrain.drop(['CallStart', 'CallEnd'], axis=1)
```

```python
cardatatest = cardatatest.drop(['CallStart', 'CallEnd'], axis=1)
```

```python
cardatatrain['Communication'] = cardatatrain['Communication'].replace({'telephone': 0, 'cellular': 1}) #replacing values in communication column, changing it to numerical value
```

```python
cardatatest['Communication'] = cardatatest['Communication'].replace({'telephone': 0, 'cellular': 1}) #replacing values in communication column, changing it to numerical value
```

```python
#replacing values - changing categorical values it to numerical values
cardatatrain['Education'] = cardatatrain['Education'].replace({'primary': 0, 'secondary': 1, 'tertiary': 2})
cardatatrain['LastContactMonth'] = cardatatrain['LastContactMonth'].replace({'jan': 0, 'feb': 1, 'mar': 2, 'apr': 3, 'may': 4, 'jun': 5, 'jul': 6, 'aug': 7, 'sep': 8, 
                                                                             'oct': 9, 'nov': 10, 'dec': 11})
```

```python
#replacing values - changing categorical values it to numerical values
cardatatest['Education'] = cardatatest['Education'].replace({'primary': 0, 'secondary': 1, 'tertiary': 2})
cardatatest['LastContactMonth'] = cardatatest['LastContactMonth'].replace({'jan': 0, 'feb': 1, 'mar': 2, 'apr': 3, 'may': 4, 'jun': 5, 'jul': 6, 'aug': 7, 'sep': 8, 
                                                                             'oct': 9, 'nov': 10, 'dec': 11})
```

```python
cardatatrain = pd.concat([cardatatrain, pd.get_dummies(cardatatrain["Job"], prefix = "Job"), 
                      pd.get_dummies(cardatatrain["Marital"], prefix = "Marital")], axis = 1)

cardatatrain=cardatatrain.drop(["Job", "Marital"], axis = 1)
```

```python
cardatatest = pd.concat([cardatatest, pd.get_dummies(cardatatest["Job"], prefix = "Job"), 
                      pd.get_dummies(cardatatest["Marital"], prefix = "Marital")], axis = 1)

cardatatest=cardatatest.drop(["Job", "Marital"], axis = 1)
```
